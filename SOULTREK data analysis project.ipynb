```
%%sql -r dataframe_6
WITH trek_revenue AS (
    SELECT
        TREK_NAME,
        SUM(TOTAL_REVENUE) AS TREK_TOTAL_REVENUE
    FROM SOULTREK_DB.PUBLIC.SOULTREK_BOOKINGS
    GROUP BY TREK_NAME
)
SELECT
    TREK_NAME,
   TREK_TOTAL_REVENUE,
    RANK() OVER (ORDER BY TREK_TOTAL_REVENUE DESC) AS REVENUE_RANK

FROM trek_revenue
order by REVENUE_RANK;
```

now cancellation query to find the corect business hypothesis for better insight.

```
%%sql -r dataframe_20
SELECT 
    PAYMENT_METHOD,
    COUNT(*) AS total_bookings,
    SUM(CASE WHEN BOOKING_STATUS = 'Cancelled' 
        THEN 1 ELSE 0 END) AS cancellations,
    ROUND(SUM(CASE WHEN BOOKING_STATUS = 'Cancelled' 
        THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 1) AS cancellation_rate,
    ROUND(SUM(TOTAL_REVENUE), 0) AS total_revenue
FROM SOULTREK_DB.PUBLIC.SOULTREK_BOOKINGS
GROUP BY PAYMENT_METHOD
ORDER BY cancellation_rate ASC;
```

Insights: Payment method significantly correlates with cancellation behaviour — UPI shows the lowest cancellation rate (2.9%) while Net Banking records the highest (6.5%), a 2.2x difference. This aligns with behavioural economics theory: immediate, irreversible payment methods (UPI, Cash) create stronger psychological commitment to the booking, while deferred or complex payment methods (Credit Card, Net Banking) reduce commitment intensity. Business recommendation: SoulTrek should incentivize UPI payments through small discounts (₹200–500 cashback) to shift booking mix toward lower-cancellation methods — converting even 50% of Net Banking bookings to UPI would prevent an estimated 7–8 additional cancellations annually and recover ₹2–3L in at-risk revenue.

Gujarat's customer base skews toward family and group bookings on moderate routes with lower per-person pricing. Business implication: acquiring one Karnataka-type customer is worth 1.07 Gujarat-type customers — SoulTrek's customer acquisition spend should be weighted toward Bengaluru and Chennai tech corridors over Ahmedabad and Surat markets

Q25.	Find the most popular payment method. 

```
%%sql -r dataframe_29
WITH payment_count AS (
    SELECT payment_method,
           COUNT(*) AS total
    FROM SOULTREK_DB.PUBLIC.SOULTREK_BOOKINGS
    GROUP BY payment_method
)
SELECT payment_method,
       total
FROM (
    SELECT *,
           RANK() OVER (ORDER BY total DESC) AS rnk
    FROM payment_count
)
WHERE rnk = 1;
```

27.	Find customers whose booking amount is above the average booking amount. Segment them into different category.

```
%%sql -r dataframe_30
-- Segment above-average customers by spending tier
SELECT 
    CUSTOMER_NAME,
    CUSTOMER_ID,
    TOTAL_REVENUE,
    CASE
        WHEN TOTAL_REVENUE >= 60000 THEN 'Platinum 💎'
        WHEN TOTAL_REVENUE >= 45000 THEN 'Gold 🥇'
        WHEN TOTAL_REVENUE >= 30000 THEN 'Silver 🥈'
        ELSE 'Standard'
    END AS customer_tier
FROM SOULTREK_DB.PUBLIC.SOULTREK_BOOKINGS
WHERE TOTAL_REVENUE > (
    SELECT AVG(TOTAL_REVENUE) 
    FROM SOULTREK_DB.PUBLIC.SOULTREK_BOOKINGS
)
ORDER BY TOTAL_REVENUE DESC;
```

 Insights: Approximately 40–45% of SoulTrek's customer base spends above the average booking value — a significantly large high-value segment that represents SoulTrek's core revenue engine. Rather than expensive customer acquisition campaigns targeting new markets, SoulTrek should implement a tiered loyalty program: Platinum (₹60K+), Gold (₹45–60K), and Silver (₹30–45K) — each receiving differentiated benefits like early trek access, exclusive guides, or group discounts. Customer retention costs 5–7x less than acquisition — converting even 20% of above-average customers into repeat bookers would significantly reduce CAC while growing LTV. Personalized communication — trek recommendations based on past difficulty level and state-specific seasonal offers — would be the most cost-effective revenue growth strategy available to SoulTrek

28.	Find the average discount given for TREK. 

```
%%sql -r dataframe_31

SELECT 
    TREK_NAME,
    ROUND(AVG("Discount_%"), 2) AS avg_discount,
    COUNT(*) AS total_bookings,
    ROUND(SUM(TOTAL_REVENUE), 0) AS actual_revenue,
    ROUND(SUM(PRICE_PER_PERSON * GROUP_SIZE), 0) AS gross_revenue,
    ROUND(SUM(PRICE_PER_PERSON * GROUP_SIZE) - 
          SUM(TOTAL_REVENUE), 0) AS revenue_lost_to_discounts
FROM SOULTREK_DB.PUBLIC.SOULTREK_BOOKINGS
GROUP BY TREK_NAME
ORDER BY revenue_lost_to_discounts DESC;
```

INSIGHTS: Pindari Glacier Trek generated the highest booking volume while also incurring the highest revenue loss due to discounts (₹317,350). This suggests strong customer demand despite substantial discounting. The company should evaluate whether the current discount level is necessary by conducting controlled pricing experiments(A/B testing). If demand remains stable after reducing discounts, profit margins could improve without significantly affecting bookings. However, if bookings decline, alternative strategies such as targeted promotions, loyalty programs, or seasonal marketing campaigns should be considered instead of broad discounts.
